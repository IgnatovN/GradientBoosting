### Let's compare performance of my Gradient Boosting Regressor implementation with Gradient Boosting Regressor from sklearn

```python
import pandas as pd

data = pd.read_csv('data.csv')
data
```

```python
from sklearn.ensemble import GradientBoostingRegressor as sklearn_regressor
from gradient_boosting import GradientBoostingRegressor as my_regressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import pandas as pd

X, y = data.drop('delay_days', axis=1), data['delay_days']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

sklearn_regr = sklearn_regressor(
    loss='squared_error',
    learning_rate=0.1,
    n_estimators=100,
    max_depth=3,
    min_samples_split=2)
self_regr =  my_regressor(
    loss="mse",
    learning_rate=0.1,
    n_estimators=100,
    max_depth=3,
    min_samples_split=2)
```

```python
%%time
sklearn_regr.fit(X_train, y_train)
sklearn_pred = sklearn_regr.predict(X_test)
```

```python
%%time 
self_regr.fit(X_train, y_train)
self_pred = self_regr.predict(X_test)
```

```python
print('Sklearn regressor MSE:', mean_squared_error(y_test, sklearn_pred, squared=True))
print('My regressor MSE: ', mean_squared_error(y_test, self_pred, squared=True))
```

#### Despite the fact, Gradient Boosting Regressor from the DS tool works much more faster, accuracy of my implementation is a little bit better
